### Metrica a utilizar

Las métricas para evaluar el desempeño de modelos en el contexto de riesgo de crédito son:

* **Recall**: Es la proporción de verdaderos positivos identificados correctamente entre todas las situaciones realmente positivas. Es crucial cuando las consecuencias de los falsos negativos son graves. En el riesgo de crédito, esto se traduce en identificar efectivamente a aquellos clientes con alta probabilidad de impago.

* **Precision**: Esta métrica evalúa cuántos de los clientes identificados por el modelo como de alto riesgo realmente presentan ese alto riesgo. En el contexto de riesgo de crédito, una alta precision indica que una gran proporción de los clientes que el modelo clasifica como de alto riesgo son efectivamente clientes con alta probabilidad de impago.

* **F1-score**: Es el promedio armónico de precisión y recall, proporcionando un equilibrio entre ambos. Es útil cuando ambos son críticos para el problema.

Teniendo en cuenta el desbalance en los datos de la variable 'credit_score', y considerando el objetivo de identificar eficientemente a los clientes de alto riesgo(clase 1), la métrica más apropiada es el **Recall para la Clase 1**. Esto ayuda a minimizar el riesgo de otorgar préstamos a aquellos más propensos al impago.

## Baseline

## Optimización del Modelo

## 1er mejor modelo: LightGBMClassifier

```python
def objective_function_lgb(trial):
    # Define the hyperparameters to tune
    params = {
        "learning_rate": trial.suggest_loguniform("learning_rate", 0.001, 0.1),
        "n_estimators": trial.suggest_int("n_estimators", 50, 1000),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "num_leaves": trial.suggest_int("num_leaves", 31, 150),  
        "min_child_samples": trial.suggest_int("min_child_samples", 20, 100),
        "reg_alpha": trial.suggest_float("reg_alpha", 0.0, 1.0),
        "reg_lambda": trial.suggest_float("reg_lambda", 0.0, 1.0),
    }

    X_train_preprocessed = preprocessor.fit_transform(X_train)
    X_val_preprocessed = preprocessor.transform(X_val)

    # Imputation of missing values
    imputer = SimpleImputer(strategy='most_frequent')
    X_train_imputed = imputer.fit_transform(X_train_preprocessed)
    X_val_imputed = imputer.transform(X_val_preprocessed)

    # Create and train the model
    model = LGBMClassifier(random_state=42, **params,
                           early_stopping_rounds=15,
                           verbosity=-1)
    model.fit(
        X_train_imputed, y_train,
        eval_set=[(X_val_imputed, y_val)],
        callbacks=[optuna.integration.LightGBMPruningCallback(trial, "binary_logloss")]
    )

    # Predict and evaluate the model
    y_pred = model.predict(X_val_imputed)
    recall = classification_report(y_val, y_pred, output_dict=True)['1'][metrica_obj]

    return -recall
```

```python
study = optuna.create_study(direction="minimize", sampler=TPESampler(seed=42))
study.optimize(objective_function_lgb, timeout=300, n_jobs=7)
```

### 2do mejor modelo: XGBClassifier

```python
def objective_function_xgb(trial):
    # Define the hyperparameters to tune
    params = {
        "learning_rate": trial.suggest_loguniform("learning_rate", 0.001, 0.1),
        "n_estimators": trial.suggest_int("n_estimators", 50, 1000),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 5),
        "reg_alpha": trial.suggest_float("reg_alpha", 0.0, 1.0),
        "reg_lambda": trial.suggest_float("reg_lambda", 0.0, 1.0),
    }

    X_train_preprocessed = preprocessor.fit_transform(X_train)
    X_val_preprocessed = preprocessor.transform(X_val)

    # Imputation of missing values
    imputer = SimpleImputer(strategy='most_frequent')
    X_train_imputed = imputer.fit_transform(X_train_preprocessed)
    X_val_imputed = imputer.transform(X_val_preprocessed)

    # Create and train the model
    model = XGBClassifier(random_state=42, **params,
                          early_stopping_rounds=15,)
    model.fit(
        X_train_imputed, y_train,
        eval_set=[(X_val_imputed, y_val)],
        verbose=False,
        callbacks=[optuna.integration.XGBoostPruningCallback(trial, "validation_0-logloss")]
    )

    # Predict and evaluate the model
    y_pred = model.predict(X_val_imputed)
    recall = classification_report(y_val, y_pred, output_dict=True)['1'][metrica_obj]

    return recall
```

```python
study_xgb = optuna.create_study(direction="maximize", sampler=TPESampler(seed=42))
study_xgb.optimize(objective_function_xgb, timeout=300, n_jobs=7)
```
