```python
class Profiler:
    def __init__(self, dataset):
        # Inicializar el profiler
        self.path = os.getcwd()
        self.date = datetime.datetime.now().strftime("%d-%m-%Y")
        new_folder = os.path.join(self.path, f"EDA_{self.date}")
        os.makedirs(new_folder, exist_ok=True)
        self.path = new_folder
        self.dataset = dataset

    def summarize(self, variables=None):
        """
        Este m√©todo se encarga de caracterizar las variables del dataset.
        
        Args:
            variables (list): Lista con las variables del dataset.
            En caso de que no se especifique, se toman todas las variables del dataset.
        
        Returns:
            Archivo summary.txt con la caracterizaci√≥n de las variables. Se almacena
            en la carpeta que se crea en el constructor.
        """
        if variables is None:
            variables = self.dataset.columns
        
        # Se crea el archivo summary.txt
        with open(os.path.join(self.path, "summary.txt"), "w") as summary:
            # Se reporta el tipo de las variables
            summary.write("Tipo de las variables:\n")
            summary.write(str(self.dataset.dtypes))
            summary.write("\n\n")
            
            # Se reporta el porcentaje de valores √∫nicos por variable
            summary.write("Porcentaje de valores √∫nicos por variable:\n")
            summary.write(str(self.dataset.nunique() / len(self.dataset) * 100))
            summary.write("\n\n")
            
            # Se reporta el porcentaje de valores nulos por variable
            summary.write("Porcentaje de valores nulos por variable:\n")
            summary.write(str(self.dataset.isnull().sum() / len(self.dataset) * 100))
            summary.write("\n\n")
            
            # En caso de ser variable num√©rica, se reporta la media, mediana, desviaci√≥n est√°ndar, m√≠nimo y m√°ximo
            # percentiles 25, 50, 75 y 100, adem√°s los porcentajes de valores cero, negativos y outliers.
            # En caso de encontrar anomal√≠as, se reporta en un archivo de texto.
            
            summary.write("Resumen de estad√≠sticas por variable:\n")
            for variable in variables:
                summary.write(f"Variable: {variable}\n")
                summary.write(f"Tipo: {self.dataset[variable].dtype}\n")
                summary.write(f"Media: {self.dataset[variable].mean()}\n")
                summary.write(f"Mediana: {self.dataset[variable].median()}\n")
                summary.write(f"Desviaci√≥n est√°ndar: {self.dataset[variable].std()}\n")
                summary.write(f"M√≠nimo: {self.dataset[variable].min()}\n")
                summary.write(f"Percentil 25: {self.dataset[variable].quantile(0.25)}\n")
                summary.write(f"Percentil 50: {self.dataset[variable].quantile(0.50)}\n")
                summary.write(f"Percentil 75: {self.dataset[variable].quantile(0.75)}\n")
                summary.write(f"Percentil 100: {self.dataset[variable].quantile(1)}\n")
                summary.write(f"M√°ximo: {self.dataset[variable].max()}\n")
                summary.write(f"Porcentaje de valores cero: {len(self.dataset[self.dataset[variable] == 0]) / len(self.dataset) * 100}\n")
                summary.write(f"Porcentaje de valores negativos: {len(self.dataset[self.dataset[variable] < 0]) / len(self.dataset) * 100}\n")
                summary.write(f"Porcentaje de valores outliers: {len(self.dataset[self.dataset[variable] > self.dataset[variable].quantile(0.75) * 1.5]) / len(self.dataset) * 100}\n")
                summary.write("\n")
                
                # Se reportan las anomal√≠as en un archivo de texto
                with open(os.path.join(self.path, f"anomalies_{variable}.txt"), "w") as anomalies:
                    anomalies.write(f"Valores cero: {len(self.dataset[self.dataset[variable] == 0])}\n")
                    anomalies.write(f"Valores negativos: {len(self.dataset[self.dataset[variable] < 0])}\n")
                    anomalies.write(f"Valores outliers: {len(self.dataset[self.dataset[variable] > self.dataset[variable].quantile(0.75) * 1.5])}\n")
                    anomalies.write("\n")
            
        
# Ejemplo de uso
# dataset = ...  # tu dataset
# profiler = Profiler(dataset)
# profiler.summarize()

        # Se reporta el porcentaje de valores √∫nicos por va
        
        
        
        
        
```
