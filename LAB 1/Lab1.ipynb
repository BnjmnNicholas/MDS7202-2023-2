Responda las siguientes preguntas (0.2 pts c/u):

1. ¿Qué es `Git` y cómo se relaciona con el control de versiones?

    R: Git es un sistema de control de vesiones distribuido, esto quiere decir que cada cliente (computador que no es el servidor) que tiene acceso al repositorio posee una copia completa de todos los datos, esto genera que haya redundancia entre los clientes, gracias a esto es muy dificil que se pierda algún archivo.  

2. Nombre y explique 3 diferencias entre `Git` y `Github`.

    R:  - Git es un DVCS y Github es un servicio de alojamiento basado en la web.
        - Git no tiene interfaz gráfica, se utiliza mediante lineas de comando, Github tiene interfaz gráfica.
        - Git crea un repo local para trackear cambios locales, Github es un espacio para subir copias de los repos de Git.
 
3. ¿Qué es un `hash` y cómo se representa en un `commit`? De un ejemplo.

    R: Un hash es una secuencia de texto (40 caracteres) que se utiliza como identificador, puede apuntar a blob, commit o árbol. Un ejemplo de un hash apuntando a un commit es el siguiente:

    ```
    $ git log
    commit 734713bc047d87bf7eac9674765ae793478c50d3
    Author: Scott Chacon <schacon@gmail.com>
    Date:   Fri Jan 2 18:32:33 2009 -0800

        fixed refs handling, added gc auto, updated tests
    ```


4. ¿Que elementos componen a la estructura de datos de `Git`?

    R: Los elemetos principales que componen a la estructura de datos de `Git` son los commit, blob y árbol. Los commit representan un snapshot del repositorio en un momento especifico, los blob almacenan el contenido de los archivos (un blob por archivo) y los árboles sirven de directorio para los blob y otros árboles.

5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos

    R: Para iniciar un repositorio desde cero se utiliza el comando:

    ```
     git init nombre-del-repo
    ```

    Para clonar un repositorio:

    ```
    git clone /ruta/al/repo.git
    ```

    Para hacer seguimiento a cambios en los archivos se tiene el comando **status** que permite ver el estado actual del repositorio, el comando **log** permite ver un historial con los últimos commits realizados mostrando la estructura de estos.?

6. `Git` siempre resuelve automáticamente los problemas de lineas al hacer merge entre diferentes ramas.

7. Ejecutar `git pull` sobre un repositorio es equivalente a ejecutar `git clone`.

    R: Falso, son similares en el sentido de que ambos cargan un estado del repositorio en entorno local, sin embargo git pull trae los cambios recientes de un repositorio y git clone crea una copia completa del repositorio en el entorno local. 

8. `git status` genera los mismos resultados que `git log`.

9. En `Git` es posible deshacer cambios y volver a versiones anteriores del código.

    R: Verdadero, es posible volver a una versión anterior del código siempre y cuando sea parte de algún commit. 

10. El comando `git commit` se utiliza para enviar cambios al repositorio remoto.

    R: Falso, realizar un commit permite capturar una instantanea de los archivos staged del repositorio y agregarlos al historial de cambios (todo esto en local). Para enviarlo al repositorio remoto se utiliza git push

11. `git stash` (0.25 puntos)

    R: Sirve para guardar temporalmente los cambios sin hacer commit, esto sirve para poder hacer pull (hay veces que el pull y los cambios generan conflicto) o cambiarse de rama sin perder el trabajo y sin tener que hacer commit. Para recuperar los cambios se utiliza `git pop`.

12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)

    R: Ambos son para deshacer cambios, git reset sirve para quitar archivos del staging area dejando el directorio sin cambios. Por otro lado git revert crea un nuevo commit que deshace todo los cambios realizados en algún commit en particular aplicando esto a la rama actual.
